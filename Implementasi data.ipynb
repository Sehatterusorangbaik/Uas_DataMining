Exploratory Data Analysis

Apakah data kita imbalance ? 

imbalance dataset yaitu apabila sebaran data untuk masing-masing kelas sangat berbeda secara ekstrim . Pada data breast cancer ini , kelas 0 (kanker jinak ) berjumlah 200-an ,sedangkan kelas 1 (kanker ganas) berjumlah 350 an atau 75% lebih banyak 

```python
sns.set_style('whitegrid')
sns.countplot(x=df['Cancer'],data=df,palette='RdBu_r')
```

Pemeriksaan sebaran nilai data pada tiap atribut/variabel untuk masing-masing kelas dengan box-plot

Dengan melihat apakah ada perbedaan sebaran nilai yang mencolok , kita bisa mendapatkan gambaran , atribut/variabel manakah yang secara umum kita duga bisa menjadi pembeda utam antara 2 kelas.

```python
l=list(df.columns[0:10])
for i in range(len(l)-1):
    sns.boxplot(x='Cancer',y=l[i], data=df, palette='winter')
    plt.figure()
```

Kita bisa menampilkan bagaimana hubungan antar atribut/variabel . Disini sebagai contoh , hanya akan ditampilkan pada 5 atribut pertama

```python
sns.pairplot(df, vars = ['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])
```

Namun, belum terlihat bagaimana hubungan antar atribut tadi jika berdasarkan kelasnya 

sns.pairplot(df, hue = 'Cancer', vars = ['mean radius','mean texture','mean perimeter','mean area','mean smoothness'])

Pemilihan Atribut (Feature Selection)
tidak semua atribut/variabel dapat menunjukan pemisahan prediksi jenis kanker dengan jelas

Perhatikan sebaran data pada kelas 1 dan 0.
Misalnya , dari dua plot berikut jelas bahwa nilai atribut mean area yang lebih kecil umumnya menunjukkan deteksi kanker positif(kanker ganas)

Sementara itu , tidak ada hal konkret yang dapat kita simpulkan dari plot atribut mean smoothness karena sebaran nilainya hampir data sama dan saling beririsan

```python
f,(ax1, ax2) = plt.subplots(1,2, sharey=True,figsize=(12,6))
ax1.scatter(df['mean area'],df['Cancer'])
ax1.set_title("Kanker menurut nilai atribut: mean area", fontsize=15)
ax2.scatter(df['mean smoothness'],df['Cancer'])
ax2.set_title("Kanker menurut nilai atribut: mean smoothness", fontsize=15)
```

```python
import seaborn as sns
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
df['Cancer'] = cancer['target']
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16,4))
fig.set_dpi(100)
ax = sns.heatmap(df[df["Cancer"]==1][['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness']], ax=axes[0])
ax.set_title('Kanker Ganas')
ax = sns.heatmap(df[df["Cancer"]==0][['mean radius', 'mean texture', 'mean perimeter', 'mean area', 'mean smoothness']], ax=axes[1])
ax.set_title('Sehat')

plt.show()



```
